```python
# Problem Solution in Leetcode
import pandas as pd

def sales_person(sales_person: pd.DataFrame, company: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    df_merged = company.merge(orders, on='com_id')
    red_com_df = df_merged[df_merged['name'] == 'RED']['sales_id'].unique()
    return sales_person[~sales_person['sales_id'].isin(red_com_df)][['name']]

```
