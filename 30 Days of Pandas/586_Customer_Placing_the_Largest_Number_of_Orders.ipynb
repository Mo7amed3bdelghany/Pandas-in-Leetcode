```python
import pandas as pd

def largest_orders(orders: pd.DataFrame) -> pd.DataFrame:
    res= orders.groupby('customer_number').count().reset_index()
    res=res.sort_values(by='order_number')
    return pd.DataFrame(res['customer_number']).tail(1)
```

# OR

```python
import pandas as pd

def largest_orders(orders: pd.DataFrame) -> pd.DataFrame:
    return pd.DataFrame(orders['customer_number'].mode())
```
