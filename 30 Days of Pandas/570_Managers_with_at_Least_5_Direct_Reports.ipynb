```python
import pandas as pd

def find_managers(employee: pd.DataFrame) -> pd.DataFrame:
    df = employee.groupby('managerId').count().reset_index()
    df = df[df['id']>4]
    df_merged = df.merge(employee, left_on='managerId',right_on='id')
    res = df_merged[['name_y']]
    res.columns=(['name'])
    return res
```
