```python
import pandas as pd

df_employees = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'name': ['Ahmed', 'Sara', 'Khaled']
})

df_salaries = pd.DataFrame({
    'emp_id': [1, 2, 4],
    'salary': [5000, 7000, 4000]
})

inner_res = pd.merge(df_employees, df_salaries, on='emp_id', how='inner')
outer_res = pd.merge(df_employees, df_salaries, on='emp_id', how='outer')
left_res = pd.merge(df_employees, df_salaries, on='emp_id', how='left')
right_res = pd.merge(df_employees, df_salaries, on='emp_id', how='right')


print("--- Inner Join ---\n", inner_res)
print("--- outer Join ---\n", outer_res)
print("--- left Join ---\n", left_res)
print("--- right Join ---\n", right_res)
```
